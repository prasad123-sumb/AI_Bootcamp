```python
# Sample Data Of KumbhMela 2025
days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7'  ]
visitors = np.random.randint(60000, 500000, size=7)
cities = ['Prayagraj', 'Varanasi', 'Lucknow',  'KedarNath', 'Badrinath', 'Kashi', 'Haridwar']
participants = np.random.randint(4000, 15000, size=7)
visitors
```

```python
 # dataframes
df_visitors = pd.DataFrame({'Day': days, 'Visitors': visitors})
df_participants = pd.DataFrame({'City': cities, 'Participants': participants})
df_visitors
df_participants
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df_visitors['Day'], df_visitors['Visitors'], marker='o', color='crimson')
plt.xlabel(' X_axis( Days ) ' )
plt.ylabel(' Y_axis( Number of Visitors ) ')
plt.title('Visitors During Khubh Mela Over Days')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.bar(df_participants['City'], df_participants['Participants'], color='pink')
plt.xlabel('Cities')
plt.ylabel('Number of Participants')
plt.title('Participants from Different Cities - Khubh Mela')
plt.show()
```

```python
# chart : Distribution of Participants by City
plt.figure(figsize=(30, 20))
plt.pie(df_participants['Participants'], labels=df_participants['City'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Participant Distribution by City')
plt.show()
```

```python
# Multiple Plots in One Figure
plt.figure(figsize=(15, 7))

# Subplot 1: Line Plot
plt.subplot(1, 2, 1)
plt.plot(df_visitors['Day'], df_visitors['Visitors'], marker='s', color='aqua')
plt.title('Daily Visitors')
plt.xlabel('Days')
plt.ylabel('Visitors')

# Subplot 2: Bar Chart
plt.subplot(1, 2, 2)
plt.bar(df_participants['City'], df_participants['Participants'], color='aqua')
plt.title('City-wise Participants')
plt.xlabel('Cities')
plt.ylabel('Participants')

plt.tight_layout()
plt.show()
 # Adding Multiple Lines 
plt.figure(figsize=(8, 5))
x = np.arange(1, 8)
plt.plot(x, visitors, label='Visitors Trend', linestyle='-', marker='o')
plt.plot(x, participants, label='Participants Trend', linestyle='--', marker='s')
plt.xlabel('Days/Cities Index')
plt.ylabel('Count')
plt.title('Visitors & Participants Trend - Khubh Mela')
plt.legend()
plt.grid(True)
plt.show()

# Saving a Plot
plt.figure(figsize=(7, 6))
plt.plot(df_visitors['Day'], df_visitors['Visitors'], color='aqua')
plt.title('Saved Plot Example - Visitors Trend')
plt.savefig('khubh_mela_visitors_trend.png')
plt.show()
```
